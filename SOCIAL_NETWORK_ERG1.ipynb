### Τοπολογία RGG  με ακτίνα σύνδεσης $R=0.35$

```python

random.seed(seedno)
n=42
R=0.35

# για την κατασκευή του RGG χρησιμοποιήστε την random_geometric_graph από το networkx
RGG=nx.random_geometric_graph(n,R,seed=random.seed(seedno))

#οπτικοποίηση: https://networkx.org/documentation/stable/auto_examples/drawing/plot_random_geometric_graph.html
pos = nx.get_node_attributes(RGG, 'pos')
nx.draw(RGG, pos, with_labels=True)
# find node near center (0.5,0.5)
dmin = 1
ncenter = 0
for n in pos:
    x, y = pos[n]
    d = (x - 0.5) ** 2 + (y - 0.5) ** 2
    if d < dmin:
        ncenter = n
        dmin = d

# color by path length from node near center
p = dict(nx.single_source_shortest_path_length(RGG, ncenter))

plt.figure(figsize=(8, 8))
nx.draw_networkx_edges(RGG, pos, alpha=0.4)
nx.draw_networkx_nodes(
    RGG,
    pos,
    nodelist=list(p.keys()),
    node_size=80,
    node_color=list(p.values()),
    cmap=plt.cm.Reds_r,
)

plt.xlim(-0.05, 1.05)
plt.ylim(-0.05, 1.05)
plt.axis("off")
plt.show()

```

### B1. Συντελεστής ομαδοποίησης (Clustering coefficient)

#### Για κάθε μια από τις τοπολογίες REG, RG(ER), RGG, SF(BA) και SW(WS):
#### Β1.1 Να υπολογιστεί ο συντελεστής ομαδοποίησης κάθε κόμβου και να αναπαρασταθεί η κατανομή του.
#### Β1.2  Να υπολογιστεί ο μέσος συντελεστής ομαδοποίησης και η διασπορά του.
#### Β.1.3 Να βρεθούν οι κόμβοι με το μικρότερο και το μεγαλύτερο συντελεστή ομαδοποίησης.
#### Να σχολιαστούν τα αποτελέσματα.

```python
import numpy as np
seedno=600
R=0.35
M=150
n=42
p=0.25
s=random.seed(seedno)
def cc(G):
    clustering_coefficients=nx.clustering(G)
    nodes=list(clustering_coefficients.values())

    #calculating the average distance for each node
    mean_clustering_coefficient= np.mean(nodes,axis=0)

    #calculating the variance for each node
    var=np.mean((nodes-mean_clustering_coefficient)**2,axis=0)

    #B.1.3
    #maximum=np.argmax(nodes)
    #minimum=np.argmin(nodes)

    maximum=[]
    minimum=[]

    length=len(nodes)

    max=-1
    min=10000

    for i in range(length):
      if nodes[i]>max:
        max=nodes[i]
      if nodes[i]<min:
        min=nodes[i]



    for i in range(length):
        if nodes[i]==max:
            maximum.append(i)
        elif nodes[i]==min:
            minimum.append(i)
    #for i in np.arange(0,len(nodes),1):
      #print(f"The clustering coefficient of the node {i} is: {nodes[i]}")
    print(f"List with the clustering coefficent of each node: {clustering_coefficients}")
    print(f"The mean clustering coefficient is : {mean_clustering_coefficient}\nThe covariance of clustering coefficient: {var}\nThe nodes with the biggest clustering coefficient are: {maximum}\nThe nodes with the least clustering coefficient are: {minimum}")
    plt.hist(list(clustering_coefficients.values()), bins=10, alpha=0.7, color='b')
    plt.xlabel('Τιμές Συντελεστή Ομαδοποίησης')
    plt.ylabel('Συχνότητα')
    plt.title('Κατανομή Συντελεστή Ομαδοποίησης')
    plt.show()

#REG
REG=nx.watts_strogatz_graph(n,d,p=0,seed=random.seed())
print("\nREG Topology")
cc(REG)
#RGGER
RGGER=nx.gnm_random_graph(n, M, seed=random.seed())
print("\nRGGER Topology")
cc(RGGER)
#RGG
RGG=nx.random_geometric_graph(n,R)
print("\nRGG Topology")
cc(RGG)
#SF
SF=nx.barabasi_albert_graph(n,d)
print("\nSF Topology")
cc(SF)
#SW
SW=nx.watts_strogatz_graph(n,d,p=0.25)
print("\nSW Topology")
cc(SW)

#comments
#Με τα βάση τα αποτελέσματα που παράχθηκαν από το κώδικα που τρέξαμε έχουμε τα έξης στοιχεία:
#Οι τοπολογιες REG,RGG έχουν κόμβους με συντελεστή ομαδοποίησης από 0.5 και πάνω ,
#οι τοπολογιές RGGER, SF έχουν κόμβους με συντελεστή ομαδοποιήσης μέχρι 50% , ενώ στην  τοπολογία SW
#η κατανομη των συντελεστών ομαδοποιήσης είναι στη κλίμακα [30%-70%]
#Συνπώς:
#Οι τοπολογίες REG,RGG έχουν μέσο συντελεστή ομαδοποίησης γύρο στο 70% (66% και 75% αντίστοιχα-συνεπώς σίγουρα πάνω από 50%) το οποίο υποδεικνύει ότι  οι κόμβοι στους γράφους αυτούς,
#έχουν την τάση να σχηματίζουν πολύ πυκνές ομάδες. Αυτό σημαίνει ότι έχουν έντονη κοινωνική δομή ή ότι σχηματίζουν στενές κοινότητες.
#Οι τοπολογίες RGGER, SF έχουν μέσο συντελεστή ομαδοποίησης πολύ πιο κάτω από 50% (20%-30% αντίστοιχα), το οποίο σημαίνει ότι
#οι κόμβοι στους γράφους αυτούς  έχουν μικρότερη τάση να σχηματίζουν πυκνές ομάδες. Αυτό μπορεί να υποδηλώνει έναν γράφο με μικρότερη κοινωνική δομή ή
# ένα δίκτυο με στενές κοινότητες. Συνήθως, αυτό μπορεί να είναι χαρακτηριστικό ενός πιο "ασυνδεδεμένου" γράφου, όπου οι κόμβοι δεν σχηματίζουν τόσο έντονες ομάδες.
# Στην τοπολογία SW, ο μέσος συντελεστής ομαδοποίησης είναι γύρω στο 50%, αυτό υποδηλώνει ότι οι κόμβοι στο γράφο αυτό έχουν τάση να σχηματίζουν σχετικά κλειστές ομάδες ή "κοινότητες".

#Όσον αφορά τωρα τη διασπορά του συντελεστή ομαδοποίησης έχουμε ότι όσο χαμηλότερη είναι η διασπορά, τόσο πιο ομοιόμορφες είναι οι τιμές,
#υποδηλώνοντας ότι οι κόμβοι έχουν παρόμοιες τιμές συντελεστή ομαδοποίησης.
#Σε όλες τις περιπτώσεις η διασπορά του συντελεστή ομαδοποίησης είναι πάρα πολύ χαμηλή και συνεπώς φαίνεται πως υπάρχει
#μια ομοιόμορφη δομή στο σχηματισμό τριγώνων ή άλλων υποδομών στο δίκτυο.
#Εξαίρεση αποτελεί η τοπολογία REG, που εξαιτίας του γεγονότος ότι όλοι οι κόμβοι της έχουν τον ίδιο συντελεστή ομαδοποίησης
#έχει διασπορά συντελεστή ομαδοποίησης =0.
#Αυτό υποδηλώνει ένα πολύ ομοιογενές δίκτυο, όπου όλοι οι κόμβοι έχουν παρόμοια περιβάλλοντα ή σχέσεις με τους γείτονές τους.

#Ανάλογα με το είδος της τοπολογίας, βλέπουμε ότι σε άλλες περιπτώσεις ο μεγαλύτερος συντελεστής ομαδοποιήσης ανήκει σε έναν κόμβο,
#και σε άλλες περιπτώσεις ο μεγαλύτερος συντελεστής ομαδοποίησης ανήκει σε 2 ή περισσότερους κόμβους.
#Στη τοπολογία REG,όπως ανεφέραμε και πιο πάνω όλοι οι κόμβοι έχουν τον ίδιο συντελεστή ομαδοποίησης,
#στις τοπολογίες RGGER και RGG : 4 κόμβοι έχουν το μεγαλύτερο συντελεστή ομαδοποίησης.
#Στις τοπολογίες SF και SW, υπάρχει μόνο ένας κόμβος που να έχει το μέγιστο συντελεστή ομαδοποίησης.
#Αυτό πρακτικά σημαίνει ότι:
#Αν ο μεγαλύτερος συντελεστής ομαδοποίησης ανήκει σε έναν κόμβο, αυτός ο κόμβος έχει ένα υψηλό επίπεδο εσωτερικής συνοχής με τους γείτονές του.
#Αν όμως ο μεγαλύτερος συντελεστής ομαδοποίησης ανήκει σε μια ομάδα κόμβων, αυτό μπορεί να υποδηλώνει ότι υπάρχει μια υπογραφή ή μια κοινότητα
#στο δίκτυο με υψηλή εσωτερική συνοχή. Αυτό μπορεί να έχει διάφορες ερμηνείες ανάλογα με το πλαίσιο της εφαρμογής.

#Σε όλες τις τοπολογίες(με εξαίρεση τη REG) υπάρχει μόνο ένας κόμβος που να έχει ελάχιστη τιμή συντελεστή ομαδοποίησης.
#Αυτό σημαίνει ότι ο κόμβος αυτός , θεωρείται "μοναχικός" και
# έχει λιγότερες συνδέσεις ή δεν είναι τόσο πυκνά συνδεδεμένος με τους γείτονές του.
```

### Β2. Μέσο μήκος μονοπατιού - Διάμετρος - Ακτίνα γράφου

#### Για κάθε μια από τις τοπολογίες REG, RG(ER), RGG, SF(BA) και SW(WS):
#### Β2.1 Nα υπολογιστεί το μέσο μήκος μονοπατιού και η διασπορά του. Προτείνεται η χρήση της συνάρτησης  all_pairs_shortest_path_length του networkx.
#### B2.2 Να υπολογιστεί η διάμετρος (συνάρτηση diameter) και η ακτίνα (συνάρτηση radius).
#### Για τα παραπάνω πραγματοποιήστε συγκρίσεις και εξηγήστε με βάση τα χαρακτηριστικά της κάθε τοπολογίας όπως παρουσιάζονται στη θεωρία.

```python
from networkx.algorithms.simple_paths import shortest_simple_paths
import numpy as np
seedno=600
n=42
R=0.35
d=4
p=0.25
M=150
def compute_path_lengths(G):
    shortest_paths=dict(nx.all_pairs_shortest_path_length(G))
    mean_path = np.mean([path for paths in shortest_paths.values() for path in paths.values()])
    var_path = np.mean([(path - mean_path)**2 for paths in shortest_paths.values() for path in paths.values()])
    diametros=nx.diameter(G)
    aktina=nx.radius(G)
    print(f"The mean distance of the path is: {mean_path}\nThe variance of the path is: {var_path}\nThe diameter of the path is:{diametros}\nThe radius of the path is:{aktina} ")



#REG
REG=nx.watts_strogatz_graph(n,d,p=0,seed=random.seed())
print("\nREG Topology")
compute_path_lengths(REG)
#RGGER
RGGER=nx.gnm_random_graph(n, M, seed=random.seed())
print("\nRGGER Topology")
compute_path_lengths(RGGER)
#RGG
RGG=nx.random_geometric_graph(n,R)
print("\nRGG Topology")
compute_path_lengths(RGG)
#SF
SF=nx.barabasi_albert_graph(n,d)
print("\nSF Topology")
compute_path_lengths(SF)
#SW
SW=nx.watts_strogatz_graph(n,d,p=0.25)
print("\nSW Topology")
compute_path_lengths(SW)


#comments
#Με βάση τους υπολογισμούς που πραγματοποιήσα η κατάταξη των τοπολογιών σε αύξουσα σειρά ως προς το συντομότερο μέσο μονοπάτι είναι:
#1.RGGER
#2.SF
#3.RGG
#4.SW
#5.REG
#Οι τοπολογίες RGGER,SF και RGG έχουν μία μικρή απόσταση από το SW. H διαφορά αυτή ανέρχεται περίπου στη 1 μονάδα.
#Αυτό είναι λογικό να συμβαίνει διότι οι συνδέσεις για τις τοπολογίες RGGER, RGG δημιουργούνται τυχαία,
#με προσαρμογή στον συνολικό αριθμό των ακμών (M) και μπορεί να προκύψουν συντομότερα μονοπάτια.
#Επίσης, στην τοπολογία Scale-Free (SF), όπου οι κόμβοι έχουν κατανομή βαθμού που ακολουθεί την κατανομή δύναμης,
#είναι λογικό να προκύπτει  σχετικά χαμηλό μέσο μονοπάτι
#Αντίθετα, στην τοπολογία Small World (SW), όπου υπάρχει μια πιθανότητα προσθήκης τυχαίων συνδέσεων για τη δημιουργία συντομότερων μονοπατιών
#μπορεί να υπάρχει λίγο μεγαλύτερο μέσο μονοπάτι.
#Η τοπολογία REG έχει πολύ μεγαλή απόσταση από τις αλλές τοπολογίες. Με βάση τη θεωρία αυτό είναι λογικό διότι,
#η τοπολογία REG δημιουργείται με τη μέθοδο Watts-Strogatz με πιθανότητα p = 0, πράγμα που σημαίνει ότι δεν υπάρχουν τυχαίες συνδέσεις και
#όλες οι ακμές συνδέονται με γειτονικούς κόμβους. Αυτό μπορεί να οδηγήσει σε μεγαλύτερα μονοπάτια από τους άλλους τύπους τοπολογιών που
#ενδέχεται να έχουν τυχαίες συνδέσεις που μειώνουν το μέσο μονοπάτι.

#Η ίδια ακριβώς κατάταξή υπάρχει για τις τοπολογίες όσον αφορά τη διάμετρο και την ακτίνα, για τους ίδιους ακριβώς λόγους.
```

### Γ4. Kεντρικότητα Katz
##### Γ4.1 Να αναπαρασταθεί η κατανομή της κεντρικότητας Katz των κόμβων για κάθε τοπολογία.
Για τον προσδιορισμό της παραμέτρου $alpha$ της συνάρτησης katz_centrality,  να υπολογιστεί η μεγαλύτερη ιδιοτιμή $λ_{max}$ του πίνακα γειτνίασης με τη συνάρτηση adjacency_spectrum του networkx.
H $alpha$ ορίζεται να είναι ιση με $\frac{1}{λ_{max}}-0.01$.
##### Γ4.2 Να υπολογιστεί η μέση  κεντρικότητα Katz  καθώς και η διασπορά της για κάθε τοπολογία.
##### Να σχολιαστούν τα αποτελέσματα.

```python
n=42
M=150
d=4
R=0.35
p=0.25
def find_katz(G):
  #calculating spectrum
  spectrum = nx.linalg.spectrum.adjacency_spectrum(G)
  lmax=max(spectrum)
  alpha=1/lmax-0.01
  katz_centralities = nx.katz_centrality(G, alpha=alpha)
  katz=list(katz_centralities.values())
  #print(katz)
  mean_katz=sum(katz)/len(katz)
  var_katz= sum((value-mean_katz)**2 for value in katz)/len(katz)

  print(f"The mean of katz centrality is: {mean_katz} and the variance of katz is: {var_katz}\n")

  plt.hist(katz_centralities.values(),bins=10, alpha=0.7, color="b")
  plt.xlabel("Katz centrality values")
  plt.ylabel("Συχνότητα")
  plt.title("Κατανομή Katz centrality")
  plt.show()

  plt.bar(katz_centralities.keys(), katz_centralities.values())
  plt.xlabel('Κόμβοι')
  plt.ylabel('Τιμές Κεντρικότητας Katz')
  plt.title('Κατανομή Κεντρικότητας Katz Κόμβων')
  plt.show()

#REG
REG=nx.watts_strogatz_graph(n,d,p=0,seed=random.seed(seedno))
print("\nREG Topology\nCalculating Katz Centrality: statistics and graphs")
find_katz(REG)
#RGGER
RGGER=nx.gnm_random_graph(n, M, seed=random.seed(seedno))
print("\nRGGER Topology\nCalculating Katz Centrality: statistics and graphs")
find_katz(RGGER)
#RGG
RGG=nx.random_geometric_graph(n,R,seed=random.seed(seedno))
print("\nRGG Topology\nCalculating Katz Centrality: statistics and graphs")
find_katz(RGG)
#SF
SF=nx.barabasi_albert_graph(n,d,seed=random.seed(seedno))
print("\nSF Topology\nCalculating Katz Centrality: statistics and graphs")
find_katz(SF)
#SW
SW=nx.watts_strogatz_graph(n,d,p=0.25,seed=random.seed(seedno))
print("\nSW Topology\nCalculating Katz Centrality: statistics and graphs")
find_katz(SW)

#comments
#Για όλες τις τοπολογίες, η κατανομή του katz centrality είναι περίπου η ίδια.
```

##### Μεταβάλλοντας την παράμετρο $g_p$ από $0$ έως $1$ με βήμα $0.1$της συνάρτησης που κατασκευάζει τοπολογίες SW μεγέθους $n=30x$ ( αντικαταστήστε το x με το τελευταίο ψηφίο του ΑΜ) με βαθμό αρχικού πλέγματος $d=4$, να βρείτε το μέσο μήκος μονοπατιού και το μέσο συντελεστή ομαδοποίησης. Πώς συμπεριφέρεται το παραγόμενο δίκτυο για $g_p = 0$, πώς για $g_p = 1$ και πώς για ενδιάμεσες τιμές $g_p \simeq 0.1 - 0.5$;

```python
random.seed(seedno)
n=302
d=4
mean_cc=0
mean_path_length=0
clustering_coefficient=0
for p in np.arange(0,1.1,0.1):
  SW=nx.watts_strogatz_graph(n,d,p,seed=random.seed(seedno))

  #calculating  the average distance for each node
  #average_shortest_path = nx.average_shortest_path_length(SW)
  shortest_paths=dict(nx.all_pairs_shortest_path_length(SW))
  mean_path = np.mean([path for paths in shortest_paths.values() for path in paths.values()])
  #print("The mean path is: ")
  #print(mean_path)
  mean_path_length+=mean_path

  #calculating the average clustering coefficient for each node
  #average_clustering = nx.average_clustering(SW)
  clustering_coefficients=nx.clustering(SW)
  nodes=list(clustering_coefficients.values())
  mean_clustering_coefficient= np.mean(nodes,axis=0)
  #print("The mean of clustering coefficient is:")
  #print(mean_clustering_coefficient)
  mean_cc+=mean_clustering_coefficient

  #printing the results
  print(f"When the value of p is {p}, the mean path is {mean_path} and the mean clustering coefficient is {mean_clustering_coefficient}\n")
print(f"\nThe mean clustering coefficient is {mean_cc/10} and the mean path is {mean_path_length/10}")
#print(mean_path_length/10)

#comments
#Για gp=0, το μέσο μήκος μονοπατιού είναι 38 και ο μέσος συντελεστής ομαδοποίησης είναι 0.5
#Για gp=1, το μέσο μήκος μονοπατιού είναι 4.35 και ο μέσος συντελεστής ομαδοποίησης είναι 0.004
#Για gp=0.1, το μέσο μήκος μονοπατιού είναι 6.46 και ο μέσος συντελεστής ομαδοποίησης είναι 0.36
#Για gp=0.2, το μέσο μήκος μονοπατιού είναι 5.66 και ο μέσος συντελεστής ομαδοποίησης είναι 0.27
#Για gp=0.3, το μέσο μήκος μονοπατιού είναι 5.03 και ο μέσος συντελεστής ομαδοποίησης είναι 0.21
#Για gp=0.4, το μέσο μήκος μονοπατιού είναι 4.85 και ο μέσος συντελεστής ομαδοποίησης είναι 0.15
#Για gp=0.5, το μέσο μήκος μονοπατιού είναι 4.45 και ο μέσος συντελεστής ομαδοποίησης είναι 0.1
#Παρατηρούμε λοιπόν ότι όσο αυξάνεται το gp τόσο μείωνεται το μέσο μήκος μονοπατιού καθώς και ο συντελεστής ομαδοποίησης.
#Όταν gp=0 μιλάμε για τοπολογία REG σε σχήμα ring network για αυτο το λόγο έχουμε σε αυτό το σημείο τέτοιες τιμές
#μέσου μήκους μονοπατιού και μέσου συντελεστή ομαδοποίησης.
#Όσο αυξάνεται το gp και μπαίνει στη κλίμακα [0.1,0.5], δημιουργούμε τοπολογία SW και όταν το gp φτάσει στο 1, έχουμε
#δημιουργήσει τοπολογία RGER
```
